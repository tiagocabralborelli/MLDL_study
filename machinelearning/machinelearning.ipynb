```python
from sklearn import datasets
```

```python
iris = datasets.load_iris()
iris.keys()
```

```python
iris.feature_names
```

```python
X = iris.data
y = iris.target
```

```python
plt.scatter(X[:,0],X[:,1], c = y)
plt.xlabel(iris.feature_names[0], fontsize = 18)
plt.ylabel(iris.feature_names[1], fontsize = 18)
plt.title("Dados da Iris",fontsize = 18)
```

```python
plt.scatter(X[:,0],X[:,2], c = y)
plt.xlabel(iris.feature_names[0], fontsize = 18)
plt.ylabel(iris.feature_names[2], fontsize = 18)
plt.title("Dados da Iris",fontsize = 18)
```

```python
plt.scatter(X[:,0],X[:,3], c = y)
plt.xlabel(iris.feature_names[0], fontsize = 18)
plt.ylabel(iris.feature_names[3], fontsize = 18)
plt.title("Dados da Iris",fontsize = 18)
```

```python
plt.hist(X[y==0,0], alpha = 0.5, label = iris.target_names[0], color = "black")
plt.hist(X[y==1,0], alpha = 0.5, label = iris.target_names[1])
plt.hist(X[y==2,0], alpha = 0.5, label = iris.target_names[2])
plt.legend()
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X,y)
gnb.predict([[3,5,4,2]])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, x_test, y_train, y_test = train_test_split(X,y)
```

```python
gnb.fit(X_train,y_train)
```

```python
y_pred_class = gnb.predict(x_test)
```

```python
from sklearn import metrics
```

```python
metrics.accuracy_score(y_test,y_pred_class)
```

```python
from sklearn.decomposition import PCA
```

```python
model = PCA(n_components=2)
model
```

```python
model.fit(X)
```

```python
model.explained_variance_ratio_
```

```python
X_2d = model.transform(X)
```

```python
plt.scatter(X_2d[:,0],X_2d[:,1], c = y)
plt.xlabel("PC1", fontsize=19)
plt.ylabel("PC2", fontsize=19)
plt.title("Anlise de Componentes Principais")
```
